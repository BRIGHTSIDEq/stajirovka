### Q1. Какие основные типы задач машинного обучения существуют?
`[Без компании, Young&&Yandex]`

**Ответ:**

Машинное обучение делится на три основных парадигмы:

**1. Обучение с учителем (Supervised Learning)**
- **Классификация** — предсказание дискретного класса (спам/не спам, отток/не отток). Алгоритмы: логистическая регрессия, SVM, деревья, бустинги, нейросети.
- **Регрессия** — предсказание непрерывной величины (цена, температура). Алгоритмы: линейная регрессия, CatBoost, нейросети.

**2. Обучение без учителя (Unsupervised Learning)**
- **Кластеризация** — группировка объектов по схожести: k-Means, DBSCAN, иерархическая кластеризация.
- **Снижение размерности** — PCA, t-SNE, UMAP, автоэнкодеры.
- **Оценка плотности** — GMM.
- **Рекомендательные системы** — матричная факторизация (ALS, SVD).

**3. Обучение с подкреплением (Reinforcement Learning)**
- Агент получает вознаграждение за действия в среде: Q-learning, Policy Gradient, PPO, A3C.

**Дополнительно:**
- **Self-supervised learning** — предобучение без разметки (BERT, GPT, MAE).
- **Semi-supervised learning** — часть данных размечена.
- **Transfer Learning** — перенос знаний из одной задачи в другую.


### Q2. Что такое переобучение, как его распознать и как бороться?
`[VK, Wildberries, Яндекс Маркет, VK]`

**Ответ:**

**Переобучение (overfitting)** — модель запомнила обучающую выборку вместо того, чтобы выявить закономерности. В результате: высокое качество на train, низкое на val/test.

**Признаки переобучения:**
- Большой gap между train loss и val loss.
- Модель идеально работает на train, но ошибается на новых данных.
- При добавлении данных качество на val растёт, а на train — падает.

**Методы борьбы:**

| Категория | Методы |
|---|---|
| **Регуляризация** | L1 (Lasso), L2 (Ridge), Elastic Net, Dropout |
| **Архитектурные** | Уменьшение сложности модели (меньше параметров) |
| **Данные** | Аугментация, сбор большего кол-ва данных |
| **Ансамбли** | Bagging (случайный лес снижает variance) |
| **Early Stopping** | Остановка обучения при росте val loss |
| **Кросс-валидация** | Надёжная оценка обобщающей способности |
| **Batch Norm** | Регуляризующий эффект |
| **Pruning** | Обрезка дерева по глубине, минимальному числу объектов |


### Q3. Метрики классификации: Accuracy, Precision, Recall, F1, ROC-AUC
`[VK, СберБанк, Ozon, Lamoda, Магнит, Rajffajzen, МТС, Яндекс Карты, ОКБ, Young&&Yandex]`

**Ответ:**

**Матрица ошибок (Confusion Matrix):**
```
               Предсказано: 1    Предсказано: 0
Реальность: 1      TP               FN
Реальность: 0      FP               TN
```

**Формулы:**
- **Accuracy** = (TP + TN) / (TP + TN + FP + FN) — доля правильных ответов. Бесполезна при дисбалансе классов.
- **Precision** (Точность) = TP / (TP + FP) — из всех предсказанных позитивных, сколько реально позитивны. Важна, когда дорого ошибиться ложно-позитивно (спам-фильтр).
- **Recall** (Полнота) = TP / (TP + FN) — из всех реальных позитивных, сколько нашли. Важна, когда дорого пропустить (диагностика рака).
- **F1** = 2 × (Precision × Recall) / (Precision + Recall) — гармоническое среднее. Баланс между precision и recall.
- **F-beta** = (1 + β²) × (Precision × Recall) / (β² × Precision + Recall). β > 1 — больший вес Recall, β < 1 — Precision.

**ROC-AUC:**
- По оси X — FPR = FP / (FP + TN), по оси Y — TPR = TP / (TP + FN) = Recall.
- AUC = площадь под ROC-кривой. Интерпретация: вероятность того, что случайно выбранный позитивный объект получит более высокий скор, чем случайно выбранный негативный.
- AUC = 0.5 → случайная модель. AUC = 1.0 → идеальная модель. AUC = 0.18 → инвертировать предсказания → AUC = 0.82!
- **ROC-AUC инвариантен к масштабированию предсказаний** (умножение на константу, возведение в степень без изменения порядка не меняет AUC).

**PR-AUC:**
- По оси X — Recall, по оси Y — Precision. Предпочтительна при сильном дисбалансе классов.


### Q4. Логистическая регрессия: формула, обучение, проверка коэффициентов
`[СберБанк, Яндекс Карты, VK, МТС, Young&&Yandex]`

**Ответ:**

**Формула:**

$$p(y=1 | x) = \sigma(w^T x + b) = \frac{1}{1 + e^{-(w^T x + b)}}$$

где σ — сигмоида. **Логит** — это $\log\frac{p}{1-p} = w^T x + b$ (линейная функция от признаков).

Знаменатель $1 + e^{-z}$: при z → +∞ → p → 1, при z → -∞ → p → 0.

**Функция потерь** — бинарная кросс-энтропия:
$$L = -\frac{1}{N}\sum_i [y_i \log(p_i) + (1-y_i) \log(1-p_i)]$$

**Обучение:** градиентный спуск. Аналитического решения нет (в отличие от линейной регрессии), т.к. функция потерь не квадратична.

**Шаг градиентного спуска:**
$$w := w - \alpha \cdot \frac{\partial L}{\partial w}$$

**Проверка значимости коэффициентов:** z-тест (коэффициент / стандартная ошибка). p-value < 0.05 → значим.

**WOE (Weight of Evidence):** монотонное преобразование категориальных признаков. WOE_i = ln(Distribution_Events_i / Distribution_Non-Events_i). Позволяет корректно использовать нелинейные зависимости в линейной модели.


### Q5. Случайный лес vs Градиентный бустинг: отличия, принципы работы
`[Без компании, VK, Lamoda, ОКБ, Сибур, Мегафон, Conundrum, Банк СТО, Райффайзен]`

**Ответ:**

**Случайный лес (Random Forest)** — **Bagging** + случайный выбор признаков:
1. Строим N деревьев, каждое — на bootstrap-выборке (случайная выборка с возвращением).
2. На каждом разбиении рассматриваем случайное подмножество признаков (√p для классификации, p/3 для регрессии).
3. Результат — усреднение (регрессия) или голосование (классификация).
4. **Снижает variance** (деревья независимы и некоррелированы). Bias почти не меняется.

**Градиентный бустинг** — **последовательный** ансамбль:
1. Инициализируем предсказание константой (среднее y для регрессии, log(p/(1-p)) для классификации).
2. Вычисляем **псевдо-остатки** (отрицательный градиент функции потерь).
3. Обучаем новое дерево предсказывать эти остатки.
4. Добавляем с шагом (learning rate): $F_m(x) = F_{m-1}(x) + \eta \cdot h_m(x)$.
5. Повторяем M раз.
6. **Снижает bias**. При большом M — риск переобучения.

**Почему не бустинг на линейных моделях?** Остатки линейной модели — тоже линейная функция от признаков, следующая линейная модель предскажет то же самое. Деревья могут аппроксимировать любые нелинейные зависимости.

| | Случайный лес | Градиентный бустинг |
|---|---|---|
| Тип ансамбля | Параллельный (Bagging) | Последовательный |
| Что снижает | Variance | Bias |
| Интерпретация | Проще | Сложнее |
| Скорость обучения | Быстрее | Медленнее |
| Устойчивость к выбросам | Выше | Ниже |
| Удаление 1 модели | Малый эффект | Малый эффект, если моделей много |


### Q6. Ключевые гиперпараметры CatBoost/XGBoost для тюнинга
`[Банк СТО]`

**Ответ:**

**Основные параметры для тюнинга:**

1. **`n_estimators` / `num_boost_round`** — число деревьев. Чем больше, тем лучше при малом learning rate, но дольше обучение.
2. **`learning_rate` / `eta`** — шаг обучения (0.01–0.3). Маленький lr → нужно больше деревьев, но лучше обобщение.
3. **`max_depth`** — глубина дерева (3–10). Больше → больше сложность → риск переобучения.
4. **`subsample`** — доля объектов на одно дерево (0.5–1.0). Стохастичность → регуляризация.
5. **`colsample_bytree`** — доля признаков на дерево. Уменьшает корреляцию деревьев.
6. **`reg_lambda` / `l2_leaf_reg`** — L2 регуляризация листьев.
7. **`min_child_samples` / `min_data_in_leaf`** — минимальное число объектов в листе. Защита от переобучения.
8. **CatBoost специфично:** `cat_features`, `border_count`, `one_hot_max_size`.
9. **Early Stopping** — остановка при отсутствии улучшения на val через `early_stopping_rounds`.

**Совет:** начинать с lr=0.1, затем снижать и увеличивать n_estimators, потом тюнить max_depth и регуляризацию.


### Q7. Регуляризация L1 vs L2: в чём разница, почему L1 зануляет веса?
`[Авито, Магнит, Lamoda, СберБанк, ДомРФ]`

**Ответ:**

**L2 (Ridge):** к функции потерь добавляется $\lambda \sum w_i^2$. Градиент: $\frac{\partial}{\partial w} \lambda w^2 = 2\lambda w$. При каждом шаге вес уменьшается пропорционально своему значению — **"weight decay"**. Веса стремятся к нулю, но никогда точно не обнуляются.

**L1 (Lasso):** добавляется $\lambda \sum |w_i|$. Градиент: $\lambda \cdot \text{sign}(w_i)$. **Постоянный шаг** в сторону нуля, не зависящий от величины веса. Поэтому L1 **обнуляет** малые веса полностью → **автоматический feature selection**.

**Геометрическая интерпретация:** область допустимых значений L2 — сфера (нет угловых точек), L1 — ромб с угловыми точками на осях координат. Решение стремится к угловой точке → некоторые координаты = 0.

**Elastic Net:** комбинация L1 + L2: $\lambda_1 \sum |w_i| + \lambda_2 \sum w_i^2$. Используется при мультиколлинеарности.

**Влияние на Bias-Variance:** регуляризация увеличивает bias, уменьшает variance. При λ = 10^6 → модель вырождается в константу (все веса → 0).


### Q8. Что такое ROC-AUC? Подробный разбор для бизнеса и технических вопросов
`[Ozon, Lamoda, Самокат, Авито, Сибур, МТС, Conundrum, Райффайзен, Young&&Yandex]`

**Ответ:**

**ROC-кривая** — кривая в пространстве (FPR, TPR) при изменении порога классификации от 1 до 0.
- **TPR (True Positive Rate) = Recall** = TP / (TP + FN). По оси Y.
- **FPR (False Positive Rate)** = FP / (FP + TN). По оси X.
- Каждый шаг вверх — объект с реальным классом 1 (шаг = 1/P).
- Каждый шаг вправо — объект с реальным классом 0 (шаг = 1/N).

**AUC** = площадь под ROC-кривой. Вероятность того, что случайный позитивный объект получит скор выше случайного негативного.

**Как объяснить бизнесу:** если у нас 100 клиентов: 50 хороших, 50 плохих — с какой вероятностью модель даст хорошему клиенту более высокий скор, чем плохому? AUC = 0.9 → с вероятностью 90%.

**Ответы на частые вопросы:**
- **Возвести скор в квадрат** → порядок ранжирования сохраняется (если скоры в [0,1]) → AUC не изменится.
- **Умножить скоры на -1** → порядок инвертируется → AUC = 1 - AUC_old.
- **Умножить на константу > 0** → порядок не меняется → AUC не изменится.
- **AUC = 0.18** → инвертировать предсказания (multiply by -1) → AUC = 0.82. Модель работает лучше случайного, просто с обратным знаком.
- **ROC-AUC при дублировании** → не изменится (порядок ранжирования не нарушается).
- **Связь с Precision/Recall:** ROC-AUC не учитывает дисбаланс, PR-AUC — учитывает. При сильном дисбалансе лучше PR-AUC.


### Q9. Метрики ранжирования: Precision@k, NDCG, MAP, Triplet Loss
`[VK]`

**Ответ:**

**Precision@k** = (релевантных среди топ-k) / k. Недостаток: не учитывает позицию в топе.

**Recall@k** = (релевантных среди топ-k) / (всего релевантных).

**MAP (Mean Average Precision)** = среднее AP по всем запросам. AP = среднее Precision@k для каждой позиции, где найден релевантный объект.

**NDCG (Normalized Discounted Cumulative Gain):**
$$DCG@k = \sum_{i=1}^{k} \frac{rel_i}{\log_2(i+1)}, \quad NDCG = \frac{DCG}{IDCG}$$
IDCG — DCG идеального ранжирования. NDCG учитывает степень релевантности (не бинарную) и позицию.

**Лоссы в ранжировании:**
- **Pointwise** — регрессия/классификация на каждый объект независимо.
- **Pairwise** — оценка пар объектов: правильно ли модель ставит релевантный выше нерелевантного. **Triplet Loss**: $L = \max(0, d(anchor, pos) - d(anchor, neg) + margin)$. Учит: anchor должен быть ближе к позитивному примеру, чем к негативному, с запасом margin.
- **Listwise** — оптимизация метрики ранжирования на всём списке (LambdaRank, LambdaMART).

**Diversity и Novelty** — дополнительные метрики рекомендательных систем. Diversity — насколько разнообразны рекомендации. Novelty — насколько новы для пользователя.


### Q10. Кросс-валидация и её особенности для временных рядов
`[Банк СТО]`

**Ответ:**

**Стандартная K-Fold CV:** данные разбиваются на K фолдов; K-1 для train, 1 для val, поворачивается K раз. **Нельзя применять к временным рядам** — происходит утечка данных из будущего в прошлое.

**Time Series Split (Walk-Forward Validation):**
- Фолд 1: train [0..t1], val [t1..t2]
- Фолд 2: train [0..t2], val [t2..t3]
- ...
- Каждый раз train расширяется, val — следующий период.

**Gap:** между train и val оставляют зазор (gap), чтобы исключить look-ahead bias при создании фич с лагами.

**В Scikit-learn:** `TimeSeriesSplit(n_splits=5, gap=7)`.

**Pipeline в Scikit-learn:** позволяет инкапсулировать последовательность преобразований + модель. Предотвращает утечку данных при CV — fit трансформеров происходит только на train фолде.


### Q11. Методы кластеризации: k-Means vs KNN, как выбрать число кластеров
`[VK, Lamoda, Сибур, Conundrum]`

**Ответ:**

**k-Means ≠ KNN:** k-Means — алгоритм кластеризации (unsupervised), KNN — классификатор/регрессор (supervised). Разные задачи.

**k-Means:**
1. Инициализируем k центроидов.
2. Каждый объект → ближайший центроид (по евклидову расстоянию).
3. Пересчитываем центроиды (среднее кластера).
4. Повторяем до сходимости.
- **Минус:** чувствителен к инициализации, предполагает сферические кластеры, нужно задать k.

**Как выбрать k:**
- **Метод локтя (Elbow Method):** строим инерцию (SSW) vs k, ищем "излом".
- **Silhouette Score:** мера компактности и разделённости кластеров ([-1, 1], чем выше — тем лучше).
- **Gap Statistic:** сравниваем с кластеризацией случайных данных.

**Другие алгоритмы:**
- **DBSCAN** — плотностная кластеризация, не нужно задавать k, находит кластеры произвольной формы, выделяет шум.
- **Hierarchical clustering** — дендрограмма, можно срезать на любом уровне.
- **GMM (Gaussian Mixture Models)** — мягкая кластеризация, вероятностная принадлежность.


### Q12. ALS и рекомендательные системы. Popularity Bias
`[Без компании, МТС]`

**Ответ:**

**ALS (Alternating Least Squares):** матричная факторизация для рекомендательных систем.
- Матрица взаимодействий R (users × items) ≈ U × V^T (низкорангового приближение).
- Оптимизация: фиксируем V → решаем аналитически для U → фиксируем U → решаем для V → чередуем.

**Заполнение матрицы для YouTube (максимизация времени просмотра):**
- r_ui = время просмотра / длительность видео (имплицитная обратная связь).
- Если просмотрено 99% — не факт что понравилось (мог смотреть по инерции). Нужно учитывать лайки, комментарии, репосты.
- Взвешенный сигнал: $r_{ui} = w_1 \cdot \text{watch\_time} + w_2 \cdot \text{like} + w_3 \cdot \text{comment} + ...$

**Восстановление матрицы через автоэнкодеры:** кодируем строку пользователя (его историю) в латентный вектор → декодируем → восстанавливаем оценки для всех айтемов.

**Popularity Bias:** популярные айтемы получают больше взаимодействий → рекомендательная система ещё активнее их рекомендует → "петля" (feedback loop). Методы борьбы:
- Downsampling популярных айтемов при обучении.
- Inverse Propensity Scoring (IPS) — взвешивание по обратной популярности.
- Отдельный exposure-aware ранжировщик.
- Exploration (ε-greedy, UCB) в продакшен-системах.

### Q13. Архитектура трансформера: подробный разбор
`[Wildberries, ГазПромНефть, СберБанк, БСВ]`

**Ответ:**

**Трансформер** состоит из стека энкодеров и декодеров (оригинал — 6+6).

**Энкодер (1 блок):**
1. **Multi-Head Self-Attention** — каждый токен смотрит на все остальные.
2. **Add & Norm** (Residual connection + Layer Norm).
3. **Feed-Forward Network** (2 линейных слоя с ReLU/GELU).
4. **Add & Norm**.

**Декодер (1 блок):**
1. **Masked Multi-Head Self-Attention** (маскируем будущие позиции — causal mask).
2. **Add & Norm**.
3. **Cross-Attention** (Query — из декодера, Key/Value — из энкодера).
4. **Add & Norm**.
5. **FFN + Add & Norm**.

**Attention формула:**
$$Attention(Q, K, V) = \text{softmax}\left(\frac{QK^T}{\sqrt{d_k}}\right) V$$

Делим на $\sqrt{d_k}$ чтобы стабилизировать градиенты (при больших $d_k$ скалярные произведения могут быть очень большими → softmax в насыщение).

**Multi-Head Attention:** запускаем h параллельных attention с разными весами → конкатенируем → линейный слой. Каждая голова учится смотреть на разные аспекты.

**Позиционное кодирование:** синусоидальные функции разных частот добавляются к эмбеддингам (трансформер не имеет рекуррентности, не знает порядка).

**Токенизация BERT:** BPE (WordPiece). Специальные токены: [CLS] (для классификации), [SEP] (разделитель), [MASK] (для MLM).

**Энкодер vs Декодер:**
- Только энкодер (двунаправленный): **BERT** — понимание текста, классификация, NER.
- Только декодер (causal): **GPT** — генерация текста.
- Энкодер + Декодер: **T5, BART** — seq2seq, перевод, суммаризация.


### Q14. Query vs Key vs Value в механизме внимания
`[СберБанк, ГазПромНефть, Wildberries]`

**Ответ:**

Аналогия: **поисковая система**.
- **Query (Q)** — запрос пользователя. Что мы ищем? (вектор текущего токена)
- **Key (K)** — индекс документов. С чем сравниваем? (векторы всех токенов)
- **Value (V)** — содержимое документов. Что берём в ответ?

**Механика:**
1. $QK^T$ → матрица схожести (attention scores) размера [seq_len × seq_len].
2. Делим на $\sqrt{d_k}$, применяем softmax → веса суммируются в 1.
3. Умножаем на V — взвешенная сумма значений.

**Q и K** оба создаются из входных эмбеддингов через разные линейные проекции ($W_Q$, $W_K$, $W_V$). Разделение Q и K позволяет модели гибко управлять похожестью: один токен может активно "искать" (Q) и пассивно "откликаться" (K) по-разному.

**Cross-attention:** Q берётся из декодера (что мы пытаемся сгенерировать), K и V — из энкодера (контекст оригинала).

### Q16. Batch Normalization: как работает, когда нужна
`[Wildberries, Райффайзен]`

**Ответ:**

**Batch Normalization** нормализует активации по батчу:
$$\hat{x} = \frac{x - \mu_B}{\sqrt{\sigma_B^2 + \epsilon}}, \quad y = \gamma \hat{x} + \beta$$

где $\mu_B$, $\sigma_B^2$ — среднее и дисперсия по батчу, $\gamma$, $\beta$ — обучаемые параметры (scale и shift).

**Зачем нужна:**
- Устраняет **Internal Covariate Shift** — смещение распределения активаций при обновлении весов.
- Позволяет использовать большие learning rate.
- Имеет **регуляризующий эффект** (шум от батча).
- Ускоряет сходимость.

**Когда применяется:** в CNN — после свёрточных слоёв (перед активацией). В трансформерах — **Layer Norm** (нормализует по размерности признаков, а не по батчу) — более стабильна.

**BN во время inference:** использует running mean/variance (накопленные во время обучения), а не батчевую статистику.

**Другие нормализации:** Layer Norm (LN) — по признакам (трансформеры), Instance Norm — по каждому объекту отдельно (стилизация), Group Norm — компромисс.


### Q17. Оптимизаторы: SGD, Momentum, Adam, RMSProp
`[СберБанк, Lamoda, Сибур]`

**Ответ:**

**SGD (Stochastic Gradient Descent):**
$$w := w - \alpha \nabla L(w)$$
Простой, но медленно сходится. **Mini-batch SGD** — компромисс между скоростью и точностью оценки градиента.

**SGD с Momentum:**
$$v := \beta v + \nabla L(w), \quad w := w - \alpha v$$
Накапливает историю градиентов → разгоняется в долинах, тормозит на осциллирующих направлениях.

**Adam (Adaptive Moment Estimation):**
$$m := \beta_1 m + (1-\beta_1)\nabla L, \quad v := \beta_2 v + (1-\beta_2)(\nabla L)^2$$
$$\hat{m} = m/(1-\beta_1^t), \quad \hat{v} = v/(1-\beta_2^t)$$
$$w := w - \alpha \hat{m} / (\sqrt{\hat{v}} + \epsilon)$$
Адаптивный lr для каждого параметра. Дефолт: $\beta_1=0.9, \beta_2=0.999$. Самый популярный оптимизатор.

**Мини-батч vs полный датасет:**
- Полный GD — медленный, точная оценка градиента, склонен к застреванию в локальных минимумах.
- SGD (батч=1) — быстрый, шумный, может выбраться из локальных минимумов.
- Mini-batch — компромисс. Большой батч → схож с GD, маленький → схож с SGD.

**Почему нельзя инициализировать веса константой?** Все нейроны в слое получат одинаковый градиент → симметрия не нарушается → нейроны учатся одному и тому же. Нужна случайная инициализация (Xavier, He).

### Q19. ЦПТ (Центральная Предельная Теорема) и ЗБЧ (Закон Больших Чисел)
`[СберБанк, Сибур]`

**Ответ:**

**ЗБЧ (Закон Больших Чисел):** при увеличении выборки выборочное среднее $\bar{X}_n$ сходится к математическому ожиданию $\mu = E[X]$. Формально (слабый ЗБЧ): $\bar{X}_n \xrightarrow{p} \mu$ при $n \to \infty$.

**ЦПТ (Центральная Предельная Теорема):** нормализованная сумма i.i.d. случайных величин с конечной дисперсией **сходится по распределению к нормальному**, независимо от распределения исходных величин:
$$\frac{\bar{X}_n - \mu}{\sigma/\sqrt{n}} \xrightarrow{d} N(0, 1)$$

**Практическое применение ЦПТ:**
- Обоснование использования z-теста и t-теста для сравнения средних.
- Базис A/B тестирования: при достаточном n метрики имеют нормальное распределение.

**Проверка нормальности:**
- Визуально: Q-Q plot, гистограмма.
- Тесты: Шапиро-Уилка (малые n), Колмогорова-Смирнова, тест Д'Агостино-Пирсона.

**Устойчивые к выбросам меры:** медиана (50-й квантиль), усечённое среднее (trimmed mean), медиана абсолютных отклонений (MAD).


### Q20. A/B тестирование: полный дизайн
`[X5, HuntIT, Lamoda, Сибур, Альфа Банк]`

**Ответ:**

**Этапы дизайна A/B теста:**

1. **Определить метрику** (ключевую + guardrail метрики). Что оптимизируем? Что нельзя ухудшить?
2. **Определить MDE** (Minimum Detectable Effect) — минимальный эффект, который хотим обнаружить.
3. **Рассчитать размер выборки:** n = (z_α + z_β)² × 2σ² / δ², где δ — MDE.
4. **Определить длительность** теста (несколько недель для seasonal effects).
5. **Рандомизация** — назначение пользователей в группы. Соль: один пользователь всегда в одной группе.
6. **A/A тест** — проверяем, что группы эквивалентны до запуска.
7. **Запуск и мониторинг** — следим за guardrail метриками.
8. **Статистический тест:** при нормальных данных — t-тест, при ненормальных — Mann-Whitney U, Bootstrap.
9. **Интерпретация:** p-value < α → отвергаем H0 (нет разницы).

**Подводные камни:**
- **Simpson's Paradox** — агрегированный эффект может быть обратным в подгруппах.
- **SUTVA violation** — пользователи контрольной и тестовой групп взаимодействуют (социальные сети).
- **Novelty effect** — первоначальный подъем, который потом спадает.
- **Multiple comparisons (Множественное сравнение)** — тестируем много метрик → повышается вероятность ложно положительных. Поправка **Бонферрони**: α_adj = α/m.
- **Мощность (Power)** = 1 - β = P(отвергнуть H0 | H1 верна). Обычно ≥ 0.8.
- **Статистическая значимость** — p-value < α (обычно 0.05). Отвергаем H0.


### Q21. SVD разложение
`[СберБанк, БСВ]`

**Ответ:**

**SVD (Singular Value Decomposition):** любую матрицу $A \in \mathbb{R}^{m \times n}$ можно разложить:
$$A = U \Sigma V^T$$

- $U \in \mathbb{R}^{m \times m}$ — матрица левых сингулярных векторов (ортогональная).
- $\Sigma \in \mathbb{R}^{m \times n}$ — диагональная матрица сингулярных значений ($\sigma_1 \geq \sigma_2 \geq ... \geq 0$).
- $V \in \mathbb{R}^{n \times n}$ — матрица правых сингулярных векторов (ортогональная).

**Любую матрицу можно разложить** (в отличие от EVD, которая требует квадратной матрицы).

**Применения:**
- **PCA** — первые k столбцов U × k наибольших σ.
- **Рекомендательные системы** — low-rank аппроксимация матрицы взаимодействий.
- **Сжатие изображений** — обрезание малых сингулярных значений.
- **Псевдообратная матрица** — решение переопределённых систем.

**Сравнение с NumPy:** np.linalg.svd даёт точное разложение (честное сравнение), но для больших матриц используют итеративные методы (ARPACK, Lanczos).


### Q22. p-value, t-тест, критические зоны
`[VK, Lamoda, МТС, Сибур, X5]`

**Ответ:**

**p-value** — вероятность получить наблюдаемую или более экстремальную статистику при условии, что нулевая гипотеза H0 верна. p < α → отвергаем H0.

**t-тест (Стьюдента):** сравнение средних двух выборок.
$$t = \frac{\bar{X}_1 - \bar{X}_2}{\sqrt{\frac{s_1^2}{n_1} + \frac{s_2^2}{n_2}}}$$

**Ограничения t-теста:** предполагает нормальность данных (или большую выборку → ЦПТ).

**При ненормальном распределении:** критерий Манна-Уитни (Wilcoxon rank-sum) — непараметрический аналог t-теста.

**Критические зоны:** область значений статистики, при которых H0 отвергается. Граница — критическое значение $t_{кр}$, соответствующее уровню значимости α.

**Что является выходом статистического теста:** t-статистика и p-value.

**Дисперсия: почему n-1?** Формула несмещённой дисперсии: $s^2 = \frac{1}{n-1}\sum(x_i - \bar{x})^2$. Используем n-1 (степени свободы), потому что мы оцениваем $\bar{x}$ из тех же данных — теряем одну степень свободы. $E[s^2] = \sigma^2$ (несмещённость).

### Q24. Словарь (dict) в Python: хэш-таблица, коллизии, рехэширование
`[Без компании, X5, Conundrum, Lamoda, СберМаркет]`

**Ответ:**

**dict** в Python — хэш-таблица с открытой адресацией (open addressing + probing).

**Принцип работы:**
1. Вызывается `hash(key)` → целое число.
2. `index = hash(key) % table_size` → ячейка в массиве.
3. Если ячейка свободна → записываем пару (key, value).
4. Если занята → **коллизия** → пробирование (quadratic probing в CPython).

**Коллизии:** одинаковый хэш у разных ключей. Решения: цепочки (chaining) или открытая адресация (в CPython используется open addressing).

**Рехэширование:** когда заполнено > 2/3 таблицы (load factor), таблица **удваивается** и все ключи пересчитываются заново. Это происходит за O(n), но амортизированная сложность вставки — O(1).

**O(1) для поиска и вставки:** в среднем случае. Worst case O(n) при многих коллизиях.

**Поиск в dict vs list:** O(1) vs O(n). Словарь даст O(1) независимо от размера, список — нужно перебирать.

**== vs is:** `==` — сравнение **значений** (`__eq__`), `is` — сравнение **идентичности объектов** (ссылок, `id()`). `None` всегда сравнивать через `is`.


### Q25. Декораторы в Python
`[ГазПромНефть, Магнит, Lamoda]`

**Ответ:**

**Декоратор** — функция (или класс), принимающая функцию и возвращающая новую функцию с расширенным поведением. Реализует паттерн **Wrapper**.


````python
import functools
import time

# Базовый декоратор
def timer(func):
    @functools.wraps(func)  # сохраняем метаданные оригинала
    def wrapper(*args, **kwargs):
        start = time.perf_counter()
        result = func(*args, **kwargs)
        elapsed = time.perf_counter() - start
        print(f"{func.__name__} выполнялась {elapsed:.4f}с")
        return result
    return wrapper

@timer
def slow_func(n):
    return sum(range(n))

slow_func(1_000_000)

# Декоратор с параметрами
def repeat(n):
    def decorator(func):
        @functools.wraps(func)
        def wrapper(*args, **kwargs):
            for _ in range(n):
                result = func(*args, **kwargs)
            return result
        return wrapper
    return decorator

@repeat(3)
def say(msg): print(msg)

say("hello")

# Стандартные декораторы из stdlib
class MyClass:
    x = 10

    @classmethod        # получает cls, не self; вызывается через класс
    def class_method(cls): return cls.x

    @staticmethod       # не получает ни self, ни cls; просто функция в пространстве имён
    def static_method(): return 42

    @property           # геттер; setter/deleter через @prop.setter
    def value(self): return self._value
````

### Q27. Ссылки в Python: copy vs deepcopy
`[ГазПромНефть, СберМаркет]`

**Ответ:**

- **Присваивание** (`b = a`) — создаёт новую ссылку на **тот же объект**. Изменение через b → видно через a.
- **`copy.copy()`** — поверхностная копия: создаётся новый объект, но вложенные объекты — те же ссылки.
- **`copy.deepcopy()`** — глубокая копия: рекурсивно копируются все вложенные объекты.


````python
import copy

# Присваивание
a = [[1, 2], [3, 4]]
b = a
b[0].append(99)
print(a)  # [[1, 2, 99], [3, 4]] — изменилось!

# Shallow copy
a = [[1, 2], [3, 4]]
b = copy.copy(a)
b[0].append(99)  # внутренний список — общий
print(a)  # [[1, 2, 99], [3, 4]]

# Deep copy
a = [[1, 2], [3, 4]]
b = copy.deepcopy(a)
b[0].append(99)  # полностью независимо
print(a)  # [[1, 2], [3, 4]] — не изменилось!
````

### Q29. Оконные функции SQL
`[Банк СТО, OZON, X5, МТС, ВТБ]`

**Ответ:**

**Оконная функция** выполняет вычисление по набору строк, связанных с текущей строкой, **не схлопывая результат** (в отличие от GROUP BY).

```sql
функция() OVER (
    [PARTITION BY column]     -- группировка (опционально)
    [ORDER BY column]         -- порядок
    [ROWS/RANGE BETWEEN ...]  -- рамка окна
)
```

**Типы функций:**
- **Агрегатные:** `SUM()`, `AVG()`, `COUNT()`, `MAX()`, `MIN()` — с OVER.
- **Ранжирующие:** `ROW_NUMBER()` (уникальный номер), `RANK()` (с пропусками при равных), `DENSE_RANK()` (без пропусков), `NTILE(n)`.
- **Смещение:** `LAG(col, n)` — значение n строк назад, `LEAD(col, n)` — вперёд, `FIRST_VALUE()`, `LAST_VALUE()`.
- **Кумулятивные:** `SUM(...) OVER (ORDER BY date ROWS UNBOUNDED PRECEDING)`.

**HAVING vs WHERE:** `WHERE` фильтрует строки **до** группировки. `HAVING` фильтрует **после** GROUP BY. `WHERE` нельзя использовать агрегатные функции, `HAVING` — можно.

**Строчные vs столбчатые БД:** строчные (PostgreSQL, MySQL) — хранят строки вместе, эффективны для OLTP (insert/update/точечные запросы). Столбчатые (ClickHouse, Redshift, BigQuery) — хранят столбцы вместе, эффективны для OLAP (агрегация по большим данным).

**EXPLAIN/EXPLAIN ANALYZE:** показывает план выполнения запроса. ANALYZE — выполняет запрос и показывает реальное время.

### Задача P1. Количество «хороших» пар
`[Ozon]`

**Условие:** Дан список чисел. Вывести количество пар (i, j) где `nums[i] == nums[j]` и `i < j`.

Пример: `[1,1,1,1]` → 6.


````python
from collections import Counter

def count_good_pairs(nums):
    """
    Для каждого числа, встречающегося k раз, таких пар C(k, 2) = k*(k-1)//2.
    Сложность: O(n) время, O(n) память.
    """
    count = Counter(nums)
    return sum(k * (k - 1) // 2 for k in count.values())

# Тест
print(count_good_pairs([1, 1, 1, 1]))   # 6
print(count_good_pairs([1, 2, 3, 1, 1, 3]))  # 4
print(count_good_pairs([1, 1, 2, 2, 3]))     # 2
````

### Задача P2. Квадраты отсортированного массива (O(n))
`[Ozon, Tinkoff, Wildberries]`

**Условие:** Дан отсортированный по возрастанию массив. Вернуть массив квадратов, также отсортированный по возрастанию.

`[-5, -3, 0, 1, 2, 4]` → `[0, 1, 4, 9, 16, 25]`


````python
def sorted_squares(nums):
    """
    Два указателя: сравниваем |левый| и |правый|.
    Больший квадрат ставим в конец результата.
    Сложность: O(n) время, O(n) память.
    """
    n = len(nums)
    result = [0] * n
    left, right = 0, n - 1
    pos = n - 1

    while left <= right:
        if abs(nums[left]) >= abs(nums[right]):
            result[pos] = nums[left] ** 2
            left += 1
        else:
            result[pos] = nums[right] ** 2
            right -= 1
        pos -= 1

    return result

print(sorted_squares([-5, -3, 0, 1, 2, 4]))  # [0, 1, 4, 9, 16, 25]
print(sorted_squares([1, 4, 10]))              # [1, 16, 100]
print(sorted_squares([-4, -1, 0, 3, 10]))      # [0, 1, 9, 16, 100]
````

### Задача P3. Стек с поддержкой getMin() за O(1)
`[Tinkoff]`

**Условие:** Реализовать стек с операциями push, pop, getMin — все за O(1).


````python
class MinStack:
    """
    Храним два стека: основной и стек минимумов.
    При push: добавляем в оба (в min_stack только если <= текущего минимума).
    При pop: если popped == min_stack.top(), тоже pop из min_stack.
    """
    def __init__(self):
        self.stack = []
        self.min_stack = []

    def push(self, val: int) -> None:
        self.stack.append(val)
        if not self.min_stack or val <= self.min_stack[-1]:
            self.min_stack.append(val)

    def pop(self) -> int:
        val = self.stack.pop()
        if val == self.min_stack[-1]:
            self.min_stack.pop()
        return val

    def getMin(self) -> int:
        return self.min_stack[-1]


# Тест
ms = MinStack()
ms.push(-2); ms.push(0); ms.push(-3)
print(ms.getMin())   # -3
ms.pop()
print(ms.getMin())   # -2
````

### Задача P4. Диапазоны чисел
`[Young&&Yandex, VK]`

**Условие:** Дан отсортированный уникальный массив. Вернуть список диапазонов, покрывающих все числа. Диапазон [a,b] → "a->b" если a≠b, иначе "a".

`[0,1,2,4,5,7]` → `["0->2","4->5","7"]`


````python
from typing import List

def summary_ranges(nums: List[int]) -> List[str]:
    """
    Проходим по массиву, отслеживаем начало диапазона.
    Если следующий элемент не consecutive — закрываем диапазон.
    O(n) время, O(1) доп. память.
    """
    if not nums:
        return []

    result = []
    start = nums[0]

    for i in range(1, len(nums) + 1):
        # Конец диапазона: либо конец массива, либо разрыв
        if i == len(nums) or nums[i] != nums[i - 1] + 1:
            end = nums[i - 1]
            result.append(str(start) if start == end else f"{start}->{end}")
            if i < len(nums):
                start = nums[i]

    return result

print(summary_ranges([0, 1, 2, 4, 5, 7]))   # ["0->2","4->5","7"]
print(summary_ranges([0, 2, 3, 4, 6, 8, 9])) # ["0","2->4","6","8->9"]
print(summary_ranges([]))                      # []
````

### Задача P5. Группировка анаграмм
`[X5]`

**Условие:** Дан список строк. Сгруппировать анаграммы вместе.

`["eat","tea","tan","ate","nat","bat"]` → `[["bat"],["nat","tan"],["ate","eat","tea"]]`


````python
from collections import defaultdict

def group_anagrams(strs: List[str]) -> List[List[str]]:
    """
    Ключ: отсортированные буквы слова. Анаграммы дадут одинаковый ключ.
    O(n * k * log k), где n — кол-во слов, k — макс длина слова.
    """
    groups = defaultdict(list)
    for word in strs:
        key = tuple(sorted(word))  # или ''.join(sorted(word))
        groups[key].append(word)
    return list(groups.values())

print(group_anagrams(["eat","tea","tan","ate","nat","bat"]))
# [['eat', 'tea', 'ate'], ['tan', 'nat'], ['bat']]
````

### Задача P6. Медиана массива
`[Conundrum, ОКБ, Яндекс]`

**Условие:** Написать функцию нахождения медианы для входного массива чисел.


````python
def find_median(nums: List[float]) -> float:
    """
    Медиана: средний элемент отсортированного массива.
    При чётном количестве — среднее двух средних.
    O(n log n) через сортировку.
    """
    if not nums:
        raise ValueError("Пустой массив")
    sorted_nums = sorted(nums)
    n = len(sorted_nums)
    mid = n // 2
    if n % 2 == 1:
        return sorted_nums[mid]
    else:
        return (sorted_nums[mid - 1] + sorted_nums[mid]) / 2

print(find_median([3, 1, 4, 1, 5]))   # 3
print(find_median([1, 2, 3, 4]))       # 2.5

# Более эффективно: O(n) алгоритм QuickSelect
import random

def quickselect_median(nums):
    """O(n) average через quickselect."""
    nums = nums[:]
    n = len(nums)

    def quickselect(lst, k):
        pivot = random.choice(lst)
        lo = [x for x in lst if x < pivot]
        eq = [x for x in lst if x == pivot]
        hi = [x for x in lst if x > pivot]
        if k < len(lo):
            return quickselect(lo, k)
        elif k < len(lo) + len(eq):
            return pivot
        else:
            return quickselect(hi, k - len(lo) - len(eq))

    mid = n // 2
    if n % 2 == 1:
        return quickselect(nums, mid)
    return (quickselect(nums, mid - 1) + quickselect(nums, mid)) / 2
````

### Задача P7. Произведение массива кроме элемента i
`[X5, СберБанк, Без компании]`

**Условие:** Дан список чисел. Для каждой позиции i вернуть произведение всех элементов кроме i-го. Без деления, O(n).


````python
def product_except_self(nums: List[int]) -> List[int]:
    """
    Префиксные и суффиксные произведения.
    result[i] = prefix[i-1] * suffix[i+1]
    O(n) время, O(1) доп. память (не считая результата).
    """
    n = len(nums)
    result = [1] * n

    # Левый проход: result[i] = произведение всех элементов левее i
    prefix = 1
    for i in range(n):
        result[i] = prefix
        prefix *= nums[i]

    # Правый проход: умножаем на произведение всех элементов правее i
    suffix = 1
    for i in range(n - 1, -1, -1):
        result[i] *= suffix
        suffix *= nums[i]

    return result

print(product_except_self([1, 2, 3, 4]))  # [24, 12, 8, 6]
print(product_except_self([0, 1, 2]))      # [2, 0, 0]
````

### Задача P8. Простое число и удаление дубликатов
`[Профи]`

**Условие 1:** Проверить, является ли число простым.
**Условие 2:** Вернуть список без дубликатов, сохранив порядок.


````python
def is_prime(n: int) -> bool:
    """O(√n) проверка простоты."""
    if n < 2:
        return False
    if n == 2:
        return True
    if n % 2 == 0:
        return False
    for i in range(3, int(n**0.5) + 1, 2):
        if n % i == 0:
            return False
    return True

print(is_prime(2))   # True
print(is_prime(17))  # True
print(is_prime(18))  # False

def remove_duplicates_ordered(lst: list) -> list:
    """
    Сохраняем порядок оригинального списка.
    dict.fromkeys() — сохраняет порядок в Python 3.7+.
    O(n) время и память.
    """
    return list(dict.fromkeys(lst))

print(remove_duplicates_ordered([1, 2, 3, 1]))         # [1, 2, 3]
print(remove_duplicates_ordered([1, 3, 2, 1, 5, 3, 5, 1, 4]))  # [1, 3, 2, 5, 4]
````

### Задача P9. Класс Random с взвешенной выборкой
`[OZON]`

**Условие:** Написать класс Random с методом sample(), возвращающим элемент из массива с заданными вероятностями: `[("a", 0.1), ("b", 0.2), ("c", 0.7)]`.


````python
import random
import bisect

class WeightedRandom:
    """
    Предварительно строим массив кумулятивных вероятностей.
    Бинарный поиск для выбора элемента.
    O(n) инициализация, O(log n) выборка.
    """
    def __init__(self, items_with_probs):
        self.items = [item for item, _ in items_with_probs]
        probs = [p for _, p in items_with_probs]
        # Кумулятивные суммы
        self.cum_probs = []
        cumsum = 0
        for p in probs:
            cumsum += p
            self.cum_probs.append(cumsum)
        # Нормализуем на случай ошибок округления
        total = self.cum_probs[-1]
        self.cum_probs = [x / total for x in self.cum_probs]

    def sample(self):
        r = random.random()  # [0, 1)
        idx = bisect.bisect_left(self.cum_probs, r)
        return self.items[min(idx, len(self.items) - 1)]


# Тест
wr = WeightedRandom([("a", 0.1), ("b", 0.2), ("c", 0.7)])
results = [wr.sample() for _ in range(10000)]
from collections import Counter
freq = Counter(results)
print({k: round(v/10000, 2) for k, v in sorted(freq.items())})
# Ожидаемо примерно: {'a': 0.1, 'b': 0.2, 'c': 0.7}
````

### Задача P10. Максимальный год с наибольшим числом дефолтов
`[ВТБ]`

**Условие:** Дан список `[[start, end], ...]` — годы дефолта клиентов. Клиент активен в годах [start, end-1]. Найти самый ранний год с максимальным числом дефолтов.


````python
def find_max_defaults_year(defaults):
    """
    Difference array (разностный массив).
    На начало дефолта +1, на конец (end) -1.
    Кумулятивная сумма — количество дефолтов в каждый год.
    O(n + range) время.
    """
    from collections import defaultdict

    delta = defaultdict(int)
    for start, end in defaults:
        delta[start] += 1
        delta[end] -= 1  # клиент НЕ учитывается в год end

    # Сортируем по годам и считаем кумулятивную сумму
    current = 0
    max_defaults = 0
    best_year = None

    for year in sorted(delta.keys()):
        current += delta[year]
        if current > max_defaults:
            max_defaults = current
            best_year = year

    return best_year

defaults = [[1950, 1961], [1960, 1971], [1970, 1981]]
print(find_max_defaults_year(defaults))  # 1960
````

### Задача P11. Дополнительные задачи: декоратор, RLE строки, assert-практика
`[RuTube, СберБанк, Ozon]`


````python
# Написать декоратор (RuTube)
import functools

def memoize(func):
    """Кэширует результаты вызовов по аргументам."""
    cache = {}
    @functools.wraps(func)
    def wrapper(*args):
        if args not in cache:
            cache[args] = func(*args)
        return cache[args]
    return wrapper

@memoize
def fib(n):
    if n < 2: return n
    return fib(n-1) + fib(n-2)

print(fib(50))  # быстро, без рекурсивного взрыва

# ── RLE строки (СберБанк) ──────────────────────────────────────────────────
def rle_encode(s: str) -> str:
    """
    'aaabbcda' → 'a3b2c1d1a1'
    Сжатие подряд идущих символов.
    """
    if not s:
        return ""
    result = []
    count = 1
    for i in range(1, len(s)):
        if s[i] == s[i-1]:
            count += 1
        else:
            result.append(f"{s[i-1]}{count}")
            count = 1
    result.append(f"{s[-1]}{count}")
    return "".join(result)

print(rle_encode("aaabbcdafgfgg"))  # a3b2c1d1a1f1g1f1g2

# ── assert оптимизация (Ozon) ──────────────────────────────────────────────
# python main.py -O  →  __debug__ = False → все assert игнорируются
# Это НЕ хорошая практика использовать assert для валидации входных данных
# в продакшн коде (т.к. -O убирает их). Используй raise ValueError/TypeError.
def f_good(nums):
    if nums is None:
        raise ValueError("nums не может быть None")
    if not hasattr(nums, 'cool_field') or nums.cool_field != 5:
        raise ValueError("cool_field должно равняться 5")
    return nums
````

### Задача P12. Сортировка файлов по цифрам и строкам
`[Яндекс]`

**Условие:** Дан список названий файлов `['1_a.txt', 'b_2.txt', ...]`. Сортировать по убыванию цифр, по возрастанию строк.


````python
import re

def sort_files(files: List[str]) -> List[str]:
    """
    Извлекаем числа и строковые части из имён файлов.
    Сортируем: числа по убыванию, строки по возрастанию.
    """
    def parse_key(filename):
        name = filename.rsplit('.', 1)[0]  # без расширения
        numbers = [int(x) for x in re.findall(r'\d+', name)]
        strings = re.findall(r'[a-zA-Z]+', name)
        num = numbers[0] if numbers else 0
        string = strings[0] if strings else ''
        return (-num, string)  # отрицательное число → убывание

    return sorted(files, key=parse_key)

files = ['1_a.txt', 'b_2.txt', '3_c.txt', 'a_1.txt', '2_b.txt']
print(sort_files(files))
# Сортировка: сначала по убыванию числа, затем по возрастанию строки
````

````python
sql_max_interval = """
WITH order_gaps AS (
    SELECT
        o.customer_id,
        c.fio,
        o.date_time::date AS order_date,
        LAG(o.date_time::date) OVER (
            PARTITION BY o.customer_id
            ORDER BY o.date_time
        ) AS prev_order_date
    FROM "order" o
    JOIN customer c ON o.customer_id = c.customer_id
)
SELECT
    customer_id,
    fio,
    MAX(order_date - prev_order_date) AS max_interval_days
FROM order_gaps
WHERE prev_order_date IS NOT NULL  -- исключаем первый заказ (нет предыдущего)
GROUP BY customer_id, fio
ORDER BY customer_id;
"""
print(sql_max_interval)
````

### Задача ML1. Минимизация LogLoss: найти оптимальный порог
`[Яндекс Маркет]`

**Условие:** Дан датасет N нулей, M единиц. Минимизировать `LogLoss = -sum(y_i * log(p) + (1-y_i) * log(1-p))` по p.


````python
import numpy as np
from scipy.optimize import minimize_scalar

def optimal_p_analytical(N_zeros: int, M_ones: int) -> float:
    \"\"\"
    Аналитическое решение: берём производную LogLoss по p и приравниваем к 0.

    L(p) = -M*log(p) - N*log(1-p)
    dL/dp = -M/p + N/(1-p) = 0
    N*p = M*(1-p)
    N*p + M*p = M
    p = M / (M + N)

    Оптимальная p = доля единиц в датасете (MAP оценка).
    \"\"\"
    return M_ones / (N_zeros + M_ones)

def logloss_constant(p: float, N: int, M: int) -> float:
    eps = 1e-15
    p = np.clip(p, eps, 1 - eps)
    return -M * np.log(p) - N * np.log(1 - p)

N, M = 40, 60  # 60 единиц из 100
p_opt = optimal_p_analytical(N, M)
print(f"Оптимальная p = {p_opt:.4f}")  # 0.6
print(f"LogLoss при p=0.6: {logloss_constant(p_opt, N, M):.4f}")
print(f"LogLoss при p=0.5: {logloss_constant(0.5, N, M):.4f}")  # больше
````

### Задача ML2. Precision и Recall для наивной модели
`[СберБанк]`

**Условие:** 100 наблюдений, 60 дефолтных. Наивная модель: каждое наблюдение = дефолт с вероятностью 0.6. Каков средний Precision и Recall?


````python
import numpy as np

"""
Наивная модель: присваивает класс 1 с вероятностью 0.6, класс 0 с вероятностью 0.4.
При пороге 0.5 → всегда предсказывает 1.

При предсказании 1 для всех 100 объектов:
- TP = 60 (реальных дефолтных, правильно предсказаны как 1)
- FP = 40 (реальных не-дефолтных, ошибочно предсказаны как 1)
- FN = 0 (все дефолтных найдены)
- TN = 0

Precision = TP / (TP + FP) = 60 / 100 = 0.6
Recall = TP / (TP + FN) = 60 / 60 = 1.0
"""

TP, FP, FN, TN = 60, 40, 0, 0
precision = TP / (TP + FP)
recall = TP / (TP + FN)
print(f"Precision: {precision}")  # 0.6
print(f"Recall: {recall}")         # 1.0

# Для случайной модели (random with P(1)=0.6):
# В среднем: TP ≈ 60*0.6=36, FP ≈ 40*0.6=24, FN ≈ 60*0.4=24
TP, FP, FN = 36, 24, 24
precision_rand = TP / (TP + FP)
recall_rand = TP / (TP + FN)
print(f"\nСлучайная модель P(1)=0.6:")
print(f"Precision: {precision_rand:.3f}")  # 0.6 = base rate
print(f"Recall: {recall_rand:.3f}")         # 0.6
````

### Задача ML3. TF-IDF вручную
`[Авито, RuTube]`

**Условие 1 (Авито):** Написать TF-IDF с нуля.
**Условие 2 (RuTube):** Посчитать TF-IDF второго предложения из текста: "Мама мыла раму. Мама любит Анну. Анна играет гамму."


````python
import math
from collections import Counter

corpus = [
    "мама мыла раму",
    "мама любит анну",
    "анна играет гамму"
]

def compute_tfidf(corpus):
    """
    TF(t,d) = count(t в d) / len(d)
    IDF(t) = log(N / df(t)), где N — кол-во документов, df — кол-во содержащих t
    TF-IDF(t,d) = TF(t,d) * IDF(t)
    """
    tokenized = [doc.split() for doc in corpus]
    N = len(tokenized)

    # DF для каждого слова
    df = Counter()
    for doc in tokenized:
        for word in set(doc):
            df[word] += 1

    # IDF
    idf = {word: math.log(N / df[word]) for word in df}

    # TF-IDF для каждого документа
    results = []
    for doc in tokenized:
        tf = Counter(doc)
        doc_len = len(doc)
        tfidf = {word: (count / doc_len) * idf[word]
                 for word, count in tf.items()}
        results.append(tfidf)

    return results, idf

tfidf_docs, idf = compute_tfidf(corpus)

print("IDF словаря:")
for word, val in sorted(idf.items()):
    print(f"  {word}: {val:.4f}")

print("\nTF-IDF второго предложения ('мама любит анну'):")
for word, val in sorted(tfidf_docs[1].items()):
    print(f"  {word}: {val:.4f}")
````

### Задача ML4. KNN с нуля (без NumPy/sklearn)
`[VK, Lamoda]`


````python
from typing import List, Tuple
import math
from collections import Counter

class KNN:
    """K-ближайших соседей на чистом Python."""

    def __init__(self, k: int = 3):
        self.k = k
        self.X_train = None
        self.y_train = None

    def fit(self, X: List[List[float]], y: List):
        self.X_train = X
        self.y_train = y

    @staticmethod
    def _euclidean(a: List[float], b: List[float]) -> float:
        return math.sqrt(sum((ai - bi) ** 2 for ai, bi in zip(a, b)))

    def predict(self, X: List[List[float]]) -> List:
        return [self._predict_single(x) for x in X]

    def _predict_single(self, x: List[float]):
        # Расстояния до всех точек обучающей выборки
        distances = [(self._euclidean(x, x_train), label)
                     for x_train, label in zip(self.X_train, self.y_train)]
        # k ближайших
        k_nearest = sorted(distances, key=lambda d: d[0])[:self.k]
        k_labels = [label for _, label in k_nearest]
        # Голосование по большинству
        return Counter(k_labels).most_common(1)[0][0]

# Тест
X_train = [[1, 2], [2, 3], [3, 3], [6, 7], [7, 8], [8, 9]]
y_train = [0, 0, 0, 1, 1, 1]

knn = KNN(k=3)
knn.fit(X_train, y_train)
print(knn.predict([[2, 2], [7, 7]]))  # [0, 1]
````

`[Ozon]`

**Условие:** В мешке 100 монет, среди них 1 фальшивая (у неё два орла). Случайно берём монету, она упала орлом. Какова вероятность, что монета фальшивая?


````python
"""
Теорема Байеса: P(F|О) = P(О|F) * P(F) / P(О)

Обозначим:
  F — монета фальшивая
  О — выпал орёл

P(F) = 1/100
P(не F) = 99/100
P(О|F) = 1 (фальшивая монета всегда орёл)
P(О|не F) = 0.5 (честная монета)

P(О) = P(О|F)*P(F) + P(О|не F)*P(не F)
     = 1 * 1/100 + 0.5 * 99/100
     = 0.01 + 0.495 = 0.505

P(F|О) = 1 * (1/100) / 0.505 ≈ 0.0198 ≈ 1/50.5
"""

P_F = 1/100
P_notF = 99/100
P_O_given_F = 1.0
P_O_given_notF = 0.5

P_O = P_O_given_F * P_F + P_O_given_notF * P_notF
P_F_given_O = (P_O_given_F * P_F) / P_O

print(f"P(О) = {P_O:.4f}")
print(f"P(Фальшивая | Орёл) = {P_F_given_O:.4f} ≈ {1/P_F_given_O:.1f}^(-1)")
````

### Задача V2. Kaggle Master (Теорема Байеса)
`[Tinkoff]`

**Условие:** 40% — Competition Master (проходят тест с P=1.0), 60% — Discussion Master (P=0.1). Кандидат прошёл тест. P(Discussion Master | прошёл тест)?


````python
P_A = 0.4   # Competition Master
P_B = 0.6   # Discussion Master
P_K_A = 1.0  # Прошёл тест | Competition Master
P_K_B = 0.1  # Прошёл тест | Discussion Master

P_K = P_A * P_K_A + P_B * P_K_B
P_B_given_K = (P_B * P_K_B) / P_K

print(f"P(K) = {P_K:.3f}")              # 0.46
print(f"P(Discussion | K) = {P_B_given_K:.4f}")  # ~0.130
print(f"P(Competition | K) = {1-P_B_given_K:.4f}")  # ~0.870
````

### Задача V3. Нечестная монета — создать честный бросок (алгоритм Фон Неймана)
`[СберБанк]`

**Условие:** Монета нечестная (P(орёл) = p ≠ 0.5, p неизвестно). Как получить справедливый (50/50) результат?


````python
import random

"""
Алгоритм фон Неймана:
1. Бросить монету дважды.
2. Если ОГ → результат: "Орёл" (p*(1-p))
3. Если ГО → результат: "Решка" (p*(1-p))
4. Если ОО или ГГ → повторить.

P(ОГ) = P(ГО) = p*(1-p) → вероятности равны → честный бросок!
"""

def biased_coin(p=0.7):
    """Нечестная монета: орёл с вероятностью p."""
    return random.random() < p

def fair_from_biased(p=0.7, max_iter=1000):
    """Возвращает 1 (орёл) или 0 (решка) с одинаковой вероятностью."""
    for _ in range(max_iter):
        r1 = biased_coin(p)
        r2 = biased_coin(p)
        if r1 and not r2:
            return 1   # Орёл-Решка → "Орёл"
        elif not r1 and r2:
            return 0   # Решка-Орёл → "Решка"
        # Иначе — повторяем
    raise RuntimeError("Превышено число итераций")

# Тест
results = [fair_from_biased(0.7) for _ in range(10000)]
print(f"P(орёл) из нечестной монеты: {sum(results)/len(results):.3f}")  # ~0.5
````

### Задача V4. Русская рулетка — вращать или нет?
`[X5]`

**Условие:** В барабане 6 гнёзд, 2 патрона **рядом**. Нажал — осечка. Нажимать снова или провернуть?


````python
"""
Расположение патронов: _ _ _ _ [P][P] (2 рядом)

После осечки мы знаем, что выстрел был из одной из 4 пустых позиций.
Пустые позиции: 1, 2, 3, 4. Патронные: 5, 6.

Если НЕ крутить (следующая позиция):
- После позиции 1 → позиция 2 (пустая) ✓
- После позиции 2 → позиция 3 (пустая) ✓
- После позиции 3 → позиция 4 (пустая) ✓
- После позиции 4 → позиция 5 (патрон!) ✗
P(выстрел | не крутить) = 1/4 = 0.25

Если КРУТИТЬ (случайная позиция):
P(выстрел | крутить) = 2/6 = 1/3 ≈ 0.333

Вывод: ЛУЧШЕ НЕ КРУТИТЬ (0.25 < 0.333)!
"""
p_shoot_no_spin = 1/4
p_shoot_spin = 2/6
print(f"P(выстрел | без вращения): {p_shoot_no_spin:.4f}")   # 0.25
print(f"P(выстрел | вращение):     {p_shoot_spin:.4f}")       # 0.333
print("Вывод: НЕ вращать барабан!")
````

### Задача V5. Вероятность пола (3 человека)
`[СберБанк]`

**Условие:** В комнате 3 человека. Какова вероятность, что хотя бы двое одного пола?


````python
"""
Вероятность что ВСЕ разного пола = 0 (нас 3 человека, полов 2 → невозможно).
Значит, всегда хотя бы двое одного пола → P = 1.0.

Альтернативная интерпретация (ровно двое одного пола):
P = 1 - P(все трое одного пола) = 1 - (1/2)^2 = 1 - 0.25 = 0.75

Формально:
P(ХМЖ или варианты) = P(2М 1Ж) + P(1М 2Ж)
P(2М 1Ж) = C(3,2) * (0.5)^3 = 3/8
P(1М 2Ж) = C(3,1) * (0.5)^3 = 3/8
P(все М) = (0.5)^3 = 1/8
P(все Ж) = (0.5)^3 = 1/8

P(хотя бы двое одного пола) = 3/8 + 3/8 + 1/8 + 1/8 = 8/8 = 1.0
"""
from math import comb

p = sum(comb(3, k) * (0.5)**3 for k in range(0, 4))
print(f"P(всех исходов) = {p}")  # 1.0 — любой исход имеет хотя бы 2 одного пола

# Ровно 2 одного пола:
p_exactly_2 = (comb(3,2) + comb(3,1)) * (0.5)**3
print(f"P(ровно 2 одного пола) = {p_exactly_2}")  # 0.75

# Хотя бы 2 одного пола (= 1.0, т.к. 3 человека, 2 пола):
print("P(хотя бы 2 одного пола) = 1.0")
````
